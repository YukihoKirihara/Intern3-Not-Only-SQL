```python
%%sql 

SELECT 
    '商学院' AS 学院,
    COUNT(CASE WHEN sex = 'Female' THEN 1 END) AS 女生申请人数,
    COUNT(CASE WHEN sex = 'Female' AND acception = 'Accepted' THEN 1 END) AS 女生录取人数,
    ROUND(COUNT(CASE WHEN sex = 'Female' AND acception = 'Accepted' THEN 1 END) / COUNT(CASE WHEN sex = 'Female' THEN 1 END) * 100, 2) AS 女生录取率,
    COUNT(CASE WHEN sex = 'Male' THEN 1 END) AS 男生申请人数,
    COUNT(CASE WHEN sex = 'Male' AND acception = 'Accepted' THEN 1 END) AS 男生录取人数,
    ROUND(COUNT(CASE WHEN sex = 'Male' AND acception = 'Accepted' THEN 1 END) / COUNT(CASE WHEN sex = 'Male' THEN 1 END) * 100, 2) AS 男生录取率,
    COUNT(*) AS 合计申请人数,
    COUNT(CASE WHEN acception = 'Accepted' THEN 1 END) AS 合计录取人数,
    ROUND(COUNT(CASE WHEN acception = 'Accepted' THEN 1 END) / COUNT(*) * 100, 2) AS 合计录取率
FROM application
WHERE department = 'Business School'

UNION

SELECT 
    '法学院' AS 学院,
    COUNT(CASE WHEN sex = 'Female' THEN 1 END) AS 女生申请人数,
    COUNT(CASE WHEN sex = 'Female' AND acception = 'Accepted' THEN 1 END) AS 女生录取人数,
    ROUND(COUNT(CASE WHEN sex = 'Female' AND acception = 'Accepted' THEN 1 END) / COUNT(CASE WHEN sex = 'Female' THEN 1 END) * 100, 2) AS 女生录取率,
    COUNT(CASE WHEN sex = 'Male' THEN 1 END) AS 男生申请人数,
    COUNT(CASE WHEN sex = 'Male' AND acception = 'Accepted' THEN 1 END) AS 男生录取人数,
    ROUND(COUNT(CASE WHEN sex = 'Male' AND acception = 'Accepted' THEN 1 END) / COUNT(CASE WHEN sex = 'Male' THEN 1 END) * 100, 2) AS 男生录取率,
    COUNT(*) AS 合计申请人数,
    COUNT(CASE WHEN acception = 'Accepted' THEN 1 END) AS 合计录取人数,
    ROUND(COUNT(CASE WHEN acception = 'Accepted' THEN 1 END) / COUNT(*) * 100, 2) AS 合计录取率
FROM application
WHERE department = 'Law School'

UNION

SELECT 
    '合计' AS 学院,
    COUNT(CASE WHEN sex = 'Female' THEN 1 END) AS 女生申请人数,
    COUNT(CASE WHEN sex = 'Female' AND acception = 'Accepted' THEN 1 END) AS 女生录取人数,
    ROUND(COUNT(CASE WHEN sex = 'Female' AND acception = 'Accepted' THEN 1 END) / COUNT(CASE WHEN sex = 'Female' THEN 1 END) * 100, 2) AS 女生录取率,
    COUNT(CASE WHEN sex = 'Male' THEN 1 END) AS 男生申请人数,
    COUNT(CASE WHEN sex = 'Male' AND acception = 'Accepted' THEN 1 END) AS 男生录取人数,
    ROUND(COUNT(CASE WHEN sex = 'Male' AND acception = 'Accepted' THEN 1 END) / COUNT(CASE WHEN sex = 'Male' THEN 1 END) * 100, 2) AS 男生录取率,
    COUNT(*) AS 合计申请人数,
    COUNT(CASE WHEN acception = 'Accepted' THEN 1 END) AS 合计录取人数,
    ROUND(COUNT(CASE WHEN acception = 'Accepted' THEN 1 END) / COUNT(*) * 100, 2) AS 合计录取率
FROM application;
```

可以看到，无论是商学院还是法学院,男生的录取率均高于女生的录取率，但是在整体上，女生的录取率高于男生的录取率，这说明**存在**辛普森悖论。

```python
%%sql

# P(yes│男生)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Male') AS P_yes_male
FROM application
WHERE sex = 'Male' AND acception = 'Accepted';
```

```python
%%sql

# P(yes│女生)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Female') AS P_yes_female
FROM application
WHERE sex = 'Female' AND acception = 'Accepted';
```

```python
%%sql

# P(no│男生)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Male') AS P_no_male
FROM application
WHERE sex = 'Male' AND acception = 'Rejected';
```

```python
%%sql

# P(no|女生)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Female') AS P_no_female
FROM application
WHERE sex = 'Female' AND acception = 'Rejected';
```

```python
%%sql

# P(yes│<男生,商学院>)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Male' AND department = 'Business school') AS P_yes_male_business
FROM application
WHERE sex = 'Male' AND department = 'Business school' AND acception = 'Accepted';
```

```python
%%sql

# P(yes│<女生,商学院>)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Female' AND department = 'Business school') AS P_yes_female_business
FROM application
WHERE sex = 'Female' AND department = 'Business school' AND acception = 'Accepted';
```

```python
%%sql

# P(yes│<男生,法学院>)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Male' AND department = 'Law school') AS P_yes_female_business
FROM application
WHERE sex = 'Male' AND department = 'Law school' AND acception = 'Accepted';
```

```python
%%sql

# P(yes│<女生,法学院>)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Female' AND department = 'Law school') AS P_yes_female_law
FROM application
WHERE sex = 'Female' AND department = 'Law school' AND acception = 'Accepted';
```

```python
%%sql

# P(no│<男生,商学院>)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Male' AND department = 'Business school') AS P_yes_female_law
FROM application
WHERE sex = 'Male' AND department = 'Business school' AND acception = 'Rejected';
```

```python
%%sql

# P(no│<女生,商学院>) 
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Female' AND department = 'Business school') AS P_no_female_business
FROM application
WHERE sex = 'Female' AND department = 'Business school' AND acception = 'Rejected';
```

```python
%%sql

# P(no│<男生,法学院>)
SELECT COUNT(*) / (SELECT COUNT(*) FROM application WHERE sex = 'Male' AND department = 'Law school') AS P_no_male_law
FROM application
WHERE sex = 'Male' AND department = 'Law school' AND acception = 'Rejected';
```

#### 任务三：实现KNN算法

```python
%%sql

DROP TABLE IF EXISTS breast_cancer_distance;

CREATE TABLE breast_cancer_distance(
    id_from INT,
    id_to INT,
    distance DOUBLE,
    is_diagnosis BOOLEAN
)
```

```python
%%sql

# 计算测试集与训练集之间的距离
INSERT INTO breast_cancer_distance(id_from, id_to, distance, is_diagnosis)
SELECT bc1.id, bc2.id, SQRT(POW(bc1.radius_mean-bc2.radius_mean, 2) + POW(bc1.texture_mean-bc2.texture_mean, 2) + POW(bc1.perimeter_mean-bc2.perimeter_mean, 2) + 
    POW(bc1.area_mean-bc2.area_mean, 2) + POW(bc1.smoothness_mean-bc2.smoothness_mean, 2) + POW(bc1.compactness_mean-bc2.compactness_mean, 2) + POW(bc1.concavity_mean-bc2.concavity_mean, 2) 
    + POW(bc1.concave_points_mean-bc2.concave_points_mean, 2) + POW(bc1.symmetry_mean-bc2.symmetry_mean, 2) + POW(bc1.fractal_dimension_mean-bc2.fractal_dimension_mean, 2) + POW(bc1.radius_se-bc2.radius_se, 2)
    + POW(bc1.texture_se-bc2.texture_se, 2) + POW(bc1.perimeter_se-bc2.perimeter_se, 2) + POW(bc1.area_se-bc2.area_se, 2) + POW(bc1.smoothness_se-bc2.smoothness_se, 2) + POW(bc1.compactness_se-bc2.compactness_se, 2) + 
    POW(bc1.concavity_se-bc2.concavity_se, 2) + POW(bc1.concave_points_se-bc2.concave_points_se, 2) + POW(bc1.symmetry_se-bc2.symmetry_se, 2) + POW(bc1.fractal_dimension_se-bc2.fractal_dimension_se, 2) 
    + POW(bc1.radius_worst-bc2.radius_worst, 2) + POW(bc1.texture_worst-bc2.texture_worst, 2) + POW(bc1.perimeter_worst-bc2.perimeter_worst, 2) + POW(bc1.area_worst-bc2.area_worst, 2) + POW(bc1.smoothness_worst-bc2.smoothness_worst, 2) 
    + POW(bc1.compactness_worst-bc2.compactness_worst, 2) + POW(bc1.concavity_worst-bc2.concavity_worst, 2) + POW(bc1.concave_points_worst-bc2.concave_points_worst, 2) + POW(bc1.symmetry_worst-bc2.symmetry_worst, 2) + POW(bc1.fractal_dimension_worst-bc2.fractal_dimension_worst, 2) ) AS distance, bc2.diagnosis
FROM breast_cancer AS bc1, breast_cancer AS bc2
WHERE (NOT bc1.is_train) AND bc2.is_train
```

这里比较难受的一个点是因为当前的MySQL Server不支持LIMIT字句后面跟随一个变量， 因此我们后续选取K值只能手动选取。

```python
%%sql

ALTER TABLE breast_cancer
ADD COLUMN predict_true BOOLEAN;

UPDATE breast_cancer
SET predict_true = IF(diagnosis = (SELECT is_diagnosis
FROM 
        (SELECT bcd.id_to, bcd.is_diagnosis
        FROM breast_cancer_distance AS bcd
        WHERE (breast_cancer.id = bcd.id_from)
        ORDER BY distance ASC
        LIMIT 13) as t
GROUP BY is_diagnosis
ORDER BY COUNT(*) DESC
LIMIT 1), TRUE, FALSE);
```
